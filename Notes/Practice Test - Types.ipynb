# Types and Typeclasses

__Types__ are sets of values.

__Typeclasses__ are sets of types.

Types and Typleclasses are a __Kind__ of think in haskell.

At the lower conceptual levels, this all has to do with *'arity'*.

![classes.gif](img/classes.gif)

Take for instance the __Typeclass__ `Num`, which contains the types `Integer`, `Int`, `Double`, and `Float`.

### Infix Functions

```haskell
-- Observe that this function takes two arguments and returns a third
:t (+)
```

This output says "There is some type `a` in the class `Num` => the function takes two `a`'s and returns a third."

### More function types

```haskell
:t zip
```

```haskell
-- Haskell's type inference can tell that te result must be an Int
-- since `length` returns an int.
f ls = head ls + length ls
:t f
```

```haskell
dE :: Int -> Int -> Bool
dE x y = (x `div` y) * y == x
```
